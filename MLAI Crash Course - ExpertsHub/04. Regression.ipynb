## Regression

### Linear Regression

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
X = diabetes['data']
y = diabetes['target']
```

```python
print(diabetes['DESCR'])
```

```python
import pandas as pd
```

```python
pd.DataFrame(X, columns=diabetes['feature_names'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = \
                train_test_split(X, y, test_size=0.2)
```

#### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_train, y_train)
```

```python
y_pred = lin_reg.predict(X_test)
```

#### Compare Pred with Actual Target

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
rmse = np.sqrt(mean_squared_error(y_pred, y_test))
```

```python
rmse
```

### Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=10)
```

```python
X_train_poly = poly.fit_transform(X_train)
```

```python
poly_reg = LinearRegression()
```

```python
poly_reg.fit(X_train_poly, y_train)
```

```python
X_test_poly = poly.fit_transform(X_test)
```

```python
y_pred_poly = poly_reg.predict(X_test_poly)
```

```python
rmse_poly = np.sqrt(mean_squared_error(y_pred_poly, y_test))
```

```python
rmse_poly
```

#### Stochastic Gradient Descent Regressor

```python
from sklearn.linear_model import SGDRegressor
```

```python
sgd_reg = SGDRegressor(tol=np.infty)
```

```python
sgd_reg.fit(X_train, y_train)
```

```python
y_pred = sgd_reg.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_pred, y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import LinearSVR
from sklearn.svm import SVR
```

```python
dt_reg = DecisionTreeRegressor()
rf_reg = RandomForestRegressor(n_estimators=100)
lin_svr_reg = LinearSVR()
svr_reg = SVR(gamma='auto')
```

```python
rmse_scores = []
for clf_ in (dt_reg, rf_reg, lin_svr_reg, svr_reg):
    clf_.fit(X_train, y_train)
    pred = clf_.predict(X_test)
    rmse_scores.append(np.sqrt(
        mean_squared_error(pred, y_test)
    ))
```

```python
rmse_scores
```
