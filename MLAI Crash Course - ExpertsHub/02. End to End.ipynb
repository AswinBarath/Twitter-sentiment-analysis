### Splitting the data
* Training Set (0.8)
* Testing Set (0.2)

How to split?
* Randomize our data
* Split on basis of ratio
* Slice array based on the ratio

```python
import numpy as np
```

```python
def split_train_test(X, y=None, test_ratio=0.2):
    random_ix = np.random.permutation(len(X))
    
    test_size = int(test_ratio * len(X))
    
    slice_ix_test = random_ix[:test_size]
    slice_ix_train = random_ix[test_size:]
    
    return X.iloc[slice_ix_train], X.iloc[slice_ix_test]
```

## Creating Custom Transformers
1. Work with Numpy arrays
    * access rows, cols by __index__
    * create variables for each col holding the index
    * combine the cols
    * return the combined Numpy array

```python
housing.columns
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.total_rooms_ix = 3
        self.total_bedrooms_ix = 4
        self.population_ix = 5
        self.households_ix = 6
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        tot_rooms_per_hhold = \
        X[:, self.total_rooms_ix] / X[:, self.households_ix]
        
        tot_bedrooms_per_hhold = \
        X[:, self.total_bedrooms_ix] / X[:, self.households_ix]
        
        pop_per_hhold = \
        X[:, self.population_ix] / X[:, self.households_ix]
        
        return np.c_[
            X,
            tot_rooms_per_hhold,
            tot_bedrooms_per_hhold,
            pop_per_hhold
        ]
```

```python
caa = CombinedAttributesAdder()
```

```python
housing_combined = caa.fit_transform(housing.values)
```
